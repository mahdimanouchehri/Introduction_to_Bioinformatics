Problem

An alignment of two strings s
and t is defined by two strings s′ and t′ satisfying the following three conditions: 1. s′ and t′ must be formed from adding gap symbols "-" to each of s and t, respectively; as a result, s and t will form subsequences of s′ and t′. 2. s′ and t′ must have the same length. 3. Two gap symbols may not be aligned; that is, if s′[j] is a gap symbol, then t′[j]

cannot be a gap symbol, and vice-versa.

We say that s′
and t′ augment s and t. Writing s′ directly over t′ so that symbols are aligned provides us with a scenario for transforming s into t. Mismatched symbols from s and t correspond to symbol substitutions; a gap symbol s′[j] aligned with a non-gap symbol t′[j] implies the insertion of this symbol into t; a gap symbol t′[j] aligned with a non-gap symbol s′[j] implies the deletion of this symbol from s

.

Thus, an alignment represents a transformation of s
into t via edit operations. We define the corresponding edit alignment score of s′ and t′ as dH(s′,t′) (Hamming distance is used because the gap symbol has been introduced for insertions and deletions). It follows that dE(s,t)=mins′,t′dH(s′,t′), where the minimum is taken over all alignments of s and t

. We call such a minimum score alignment an optimal alignment (with respect to edit distance).

Given: Two protein strings s
and t

in FASTA format (with each string having length at most 1000 aa).

Return: The edit distance dE(s,t)
followed by two augmented strings s′ and t′ representing an optimal alignment of s and t.

```python
x=''
header=''
y=''
fasta = {}
with open('rosalind_edta.txt','r') as f :
    for line in f:
        line = line.strip()
        if not line:
            continue
        if line.startswith(">"):
            header = line[1:]
            if header not in fasta:
                fasta[header] = []
            continue
        sequence = line
        fasta[header].append(sequence)
a=list(fasta)
for i in fasta[a[0]]:
    x=x+i
for i in fasta[a[1]]:
    y=y+i

x = list(x)

y=list(y)

m = len(x)
n = len(y)
map = []
for i in range(m+1):
    map.append([])
    for j in range(n+1):
        map[i].append(0)
for i in range (m+1):
    map[i][0]=i*1
for i in range (n+1):
    map[0][i]=i*1

for i in range (1,m+1):
    for j in range (1,n+1):
        if x[i-1]==y[j-1]:
            map[i][j]=map[i-1][j-1]
        else:
            map[i][j]=min(map[i-1][j-1]+1,map[i-1][j]+1,map[i][j-1]+1)

l=n+m
i=m
j=n
xpos=l
ypos=l
xans=[0]*(l+1)
yans=[0]*(l+1)
while  i != 0 and j != 0 :
    if x[i-1]==y[j-1]:
        xans[xpos]=x[i-1]
        xpos =xpos-1
        yans[ypos] = y[j - 1]
        ypos =ypos-1
        i=i-1
        j=j-1
        continue
    elif map[i-1][j-1]+1==map[i][j]:
        xans[xpos] = x[i - 1]
        xpos = xpos - 1
        yans[ypos] = y[j - 1]
        ypos = ypos - 1
        i = i - 1
        j = j - 1
        continue
    elif map[i-1][j]+1==map[i][j]:
        xans[xpos] = x[i - 1]
        xpos = xpos - 1
        yans[ypos] = '-'
        ypos = ypos - 1
        i=i-1
        continue
    elif map[i][j-1]+1==map[i][j]:
        xans[xpos] = '-'
        xpos = xpos - 1
        yans[ypos] = y[j - 1]
        ypos = ypos - 1
        j=j-1
        

while xpos > 0 :
    if i>0 :
        xans[xpos]=x[i-1]
        xpos=xpos-1
        i=i-1
    else:
        xans[xpos]='-'
        xpos = xpos - 1

while ypos > 0 :
    if j>0 :
        yans[ypos] = y[j - 1]
        ypos=ypos-1
        j=j-1
    else:
        yans[ypos] = '-'
        ypos = ypos - 1


id=1
i=l
while True :
    if yans[i]=='-' and xans[i]=='-' :
        id = i + 1
        break
    else :
        i=i-1
print(map[m][n])
for i in range(id, l+1):
    print(xans[i], end='')

print('')
for i in range (id,l+1):
    print(yans[i],end='')
```
